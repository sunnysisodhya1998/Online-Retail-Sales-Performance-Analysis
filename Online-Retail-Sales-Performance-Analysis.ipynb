```python
data['Year']=pd.DatetimeIndex(data['InvoiceDate']).year
```

```python
Sales=data.loc[data['Quantity']>0&~(data['InvoiceNo'].str.contains('C'))]
```

```python
Sales
```

```python
Sales.sample(10)
```

```python
Sales_New=Sales.copy()
```

```python
Sales_New['Revenue']=Sales_New['Quantity']*Sales_New['UnitPrice']
```

```python
Sales_New.head()
```

```python
Sales_Mean=Sales_New.groupby('Year').mean()['Revenue']
```

```python
Sales_Mean.head()
```

```python
Sales_Mean=Sales_Mean.reset_index()
```

```python
Sales_Mean.head()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(Sales_Mean['Year'], Sales_Mean['Revenue'],marker='o')
plt.title('Average of Revenue per Year',fontsize=16)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Average of Revenue', fontsize = 12)
plt.ylabel('Average of Revenue',fontsize=12)
plt.grid(color='darkgrey',linestyle=':',linewidth=0.5)
plt.gca().set_xticks([2009,2010,2011])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```

```python
Sales_Finish=Sales_New[Sales_New['CustomerID'].notna()]
```

```python
Sales_Finish.count()
```

```python
Purchase_Canceled=data[data['InvoiceNo'].str.contains('C')]
```

```python
Purchase_Canceled.count()
```

```python
Count_Finished=Sales_Finish.groupby('Year').count()['InvoiceNo'].reset_index()
```

```python
Count_Finished
```

```python
Count_Canceled=Purchase_Canceled.groupby('Year').count()['InvoiceNo'].reset_index()
sns.lineplot(Count_Canceled['Year'],Count_Canceled['InvoiceNo'],marker='o',color='red',label='Cancel')
```

```python
Count_Canceled
```

```python
sns.lineplot(Count_Finished['Year'],Count_Finished['InvoiceNo'],marker='o',color='green',label='Finished')

sns.lineplot(Count_Canceled['Year'],Count_Canceled['InvoiceNo'],marker='o',color='red',label='Cancel')

plt.title('Finished & Cancel',fontsize=16)

plt.xlabel('Year', fontsize=12)

plt.ylabel('InvoiceNo', fontsize=12)

plt.grid(color='darkgrey',linestyle=':',linewidth=0.5)

plt.legend()

plt.gca().set_xticks([2009,2010,2011])

plt.gca().spines['top'].set_visible(False)

plt.gca().spines['right'].set_visible(False)
```

```python
Cancellation_Rate=Count_Canceled['InvoiceNo']/data['InvoiceNo'].count()*100

Cancellation=Count_Canceled.assign(Cancellation_Rate=Count_Canceled['InvoiceNo']/data['InvoiceNo'].count()*100)
Cancellation[['Year','Cancellation_Rate']]
```

```python
sns.lineplot(Cancellation['Year'],Cancellation['Cancellation_Rate'],marker='o',color='red',label='Cancel')

plt.title('Cancellation Rate',fontsize=16)

plt.xlabel('Year',fontsize=12)

plt.ylabel('Cancellation',fontsize=12)

plt.grid(color='darkgrey',linestyle=':',linewidth=0.5
                                                                            )
plt.legend()

plt.gca().set_xticks([2009,2010,2011])

plt.gca().spines['top'].set_visible(False)

plt.gca().spines['right'].set_visible(False)


```

```python
Comparison=Cancellation

Comparison['Total_Finished']=Count_Finished['InvoiceNo']

Comparison['Total_Canceled']=Comparison['InvoiceNo']

Comparison[['Year','Total_Finished','Total_Canceled','Cancellation_Rate']]
```
